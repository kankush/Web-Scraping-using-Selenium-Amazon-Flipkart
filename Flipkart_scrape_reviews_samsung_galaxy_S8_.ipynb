## Scraping Amazon Reviews

#### Import the required libraries

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import os
import csv
```

```python
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
print("Flipkart - Scraping Reviews : Model - Samsung Galaxy S8")
#print("Scraping Start time ",datetime.datetime.now())
# chrome_options = webdriver.ChromeOptions()
# driver = webdriver.Chrome('./data/chromedriver.exe',options=chrome_options)
driver = webdriver.Chrome(ChromeDriverManager().install())

dataFrame_Reviews = pd.DataFrame(columns= ['Reviewer Name','Review Title','Review Text','Review Rating'])

for i in range (1,2):
    pageBaseURL = "https://www.flipkart.com/samsung-galaxy-s8-orchid-grey-64-gb/product-reviews/itmeveqphhzvkrxy?pid=MOBEVEQPQQUZ575Y&page="
    dynamicURL = pageBaseURL + str(i)
    driver.get(dynamicURL)
    assert "Samsung" in driver.title   # confirm with assert() if 'Samsung' is in page title
    review_elements = driver.find_elements_by_css_selector("._1gY8H-") # this will give the complete review block
    print("Scraping reviews page - " + str(i)) # Print Page Number
    for review_element in review_elements:
        try:
            reviewer_name_element = review_element.find_element_by_css_selector("._3sxSiS")
            reviewer_name = reviewer_name_element.text # Reviewer Name
            
            review_title_element = review_element.find_element_by_css_selector("._2xg6Ul")
            review_title = review_title_element.text # Review Title
            
            review_text_element = review_element.find_element_by_css_selector(".qwjRop div")
            reviewer_review = review_text_element.text
            try:
                if "READ MORE" in reviewer_review:
                    readmorebutton = review_text_element.find_element_by_css_selector ("._1EPkIx span")
                    readmorebutton.click()                
                    review_text_element = review_element.find_element_by_css_selector(".qwjRop div")
                    reviewer_review = review_text_element.text
                
            except NoSuchElementException:                
                review_text_element = review_element.find_element_by_css_selector(".qwjRop div")
                reviewer_review=review_text_element.text # Review Text
                
            
            review_rating_element = review_element.find_element_by_css_selector(".E_uFuv")
            review_rating = review_rating_element.text # Review Rating
            
            review_details=[reviewer_name, review_title, reviewer_review,review_rating] # All Reivew details 

            record = [review_details[0], review_details[1],review_details[2],review_details[3]] # Get all review details as one record
            dataFrame_Reviews.loc[dataFrame_Reviews.shape[0]] = record # Add the review record to the dataframe
        except NoSuchElementException:
            pass
                    
dataFrame_Reviews.to_csv('Samsung_s8_review_scraping.csv') # Export the reivews dataframe to the csv file
#print("Scraping End time ",datetime.datetime.now()) # print the time when scrapting completed
dataFrame_Reviews.head(10) # Print top 10 rows

```
